```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Task 1: Load and Explore the Dataset
# Load the Iris dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = [data.target_names[i] for i in data.target]

# Display the first few rows
print(df.head())
```

```python
# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Task 2: Basic Data Analysis
# Compute basic statistics
print("\nBasic statistics:\n", df.describe())
```

```python
# Group by species and compute mean of numerical columns
species_mean = df.groupby('species').mean()
print("\nMean values by species:\n", species_mean)
```

```python
# Task 3: Data Visualization
plt.figure(figsize=(12, 8))

# Line Chart - Sepal Length trends
plt.subplot(2, 2, 1)
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length', color='b')
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()

# Bar Chart - Average petal length per species
plt.subplot(2, 2, 2)
sns.barplot(x=species_mean.index, y=species_mean['petal length (cm)'], palette='viridis')
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")

# Histogram - Distribution of Sepal Width
plt.subplot(2, 2, 3)
plt.hist(df['sepal width (cm)'], bins=15, color='g', alpha=0.7)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")

# Scatter Plot - Sepal Length vs Petal Length
plt.subplot(2, 2, 4)
sns.scatterplot(x=df['sepal length (cm)'], y=df['petal length (cm)'], hue=df['species'], palette='deep')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")

plt.tight_layout()
plt.show()
```

```python
# Findings
print("\nFindings:")
print("- Setosa species tend to have the smallest petal length.")
print("- The Sepal Length vs Petal Length scatter plot shows a clear separation between species.")
print("- The histogram suggests most sepal width values are clustered around 3 cm.")
```
